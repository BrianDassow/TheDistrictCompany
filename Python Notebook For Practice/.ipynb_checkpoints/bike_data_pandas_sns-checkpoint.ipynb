###### Load pyplot and seaborn libraries, then construct a pairplot and a scatter plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(style='whitegrid',context='notebook')
```

```python
cols = ['weekday','temp','hum','windspeed','cnt']
```

```python
sns.pairplot(df[cols],size=2.5);
```

```python
plt.scatter(df['temp'],df['cnt']);plt.xlabel('temp');plt.ylabel('cnt');
```

###### Construct a heatmap revealing the correlation matrix relating the variables in the previous pair plot

```python
import numpy as np
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm,cbar=True,
                annot=True,
                square=True,
                fmt='.2f',
                annot_kws={'size':15},
                yticklabels=cols,
                xticklabels=cols)
```

```python
X = df[['temp']].values
y = df['cnt'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression()
slr.fit(X, y)
y_pred = slr.predict(X)
print('Slope: %.3f' % slr.coef_[0])
print('Intercept: %.3f' % slr.intercept_)
```

```python
def lin_regplot(X, y, model):
    plt.scatter(X, y, c='lightblue')
    plt.plot(X, model.predict(X), color='red', linewidth=2)    
    return 
```

```python
lin_regplot(X, y, slr)
plt.xlabel('Temperature [temp]')
plt.ylabel('Bike Usage Count [cnt]')
plt.tight_layout()
#plt.savefig('scikit_lr_fit.png', dpi=300)
```
