```python
from pylab import *
%matplotlib inline
```

```python
seed(5)
num = 10000
x = zeros(num+1) 
sx = 0
y = zeros(num+1) 
sy = 0
```

```python
for n in range(num):
    if (round(rand())):
        sx += 1
        x[n] = sx
    else:
        sx += -1
        x[n] = sx
    if (round(rand())):
        sy += 1
        y[n] = sy
    else:
        sy += -1
        y[n] = sy
```

```python
plot(x,y,'ro')
xlim(-200,200)
ylim(-200,200)
```
