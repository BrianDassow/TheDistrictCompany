```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
import os
os.chdir("Files")
```

```python
sns.set(style='whitegrid',context='notebook')
filename = "sales.csv"
data = pd.read_csv(filename)
df = data.T
for n in range(df.shape[1]):
    df.rename(columns={df.columns[n]: df.iloc[0,n]}, inplace=True)
df = df[1:]
df.head()
```

```python
dfMon = df.ix[0::7]
my_plot3 = dfMon.plot(kind='bar',stacked=True,title="Total Sales by Customers")
my_plot3.set_xlabel("Mondays")
my_plot3.set_ylabel("Revenue")
my_plot3.legend(dfMon, loc='center left', ncol=2,bbox_to_anchor=(1.2, 0.5))
dfMon.head()
```
